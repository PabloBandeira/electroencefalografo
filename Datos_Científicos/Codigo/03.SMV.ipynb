```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Cargar los datos
verdad_data = pd.read_csv('data/TP/verdad.dat', delimiter=' ', names = ['timestamp','counter','eeg','attention','meditation','blinking'])
mentira_data = pd.read_csv('data/TP/mentira.dat', delimiter=' ', names = ['timestamp','counter','eeg','attention','meditation','blinking'])

# Agregar una columna de etiquetas (0 para verdad, 1 para mentira)
verdad_data['label'] = 0
mentira_data['label'] = 1

# Convertir las columnas a tipos numéricos
verdad_data = verdad_data.apply(pd.to_numeric, errors='coerce')
mentira_data = mentira_data.apply(pd.to_numeric, errors='coerce')

# Eliminar filas con valores nulos
verdad_data = verdad_data.dropna()
mentira_data = mentira_data.dropna()

# Concatenar los dos conjuntos de datos
combined_data = pd.concat([verdad_data, mentira_data], ignore_index=True)

combined_data
```

```python

# Dividir los datos en características (X) y etiquetas (y)
X = combined_data[['eeg', 'attention', 'meditation', 'blinking']]
y = combined_data['label']

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Crear y entrenar el modelo SVM para clasificación
model = SVC(kernel='rbf')  # Puedes ajustar el tipo de kernel según tus necesidades
model.fit(X_train, y_train)

# Hacer predicciones en el conjunto de prueba
predictions = model.predict(X_test)

# Evaluar el rendimiento del modelo
accuracy = accuracy_score(y_test, predictions)
print(f'Accuracy: {accuracy}')

# Imprimir el informe de clasificación
print('Classification Report:\n', classification_report(y_test, predictions))
```
